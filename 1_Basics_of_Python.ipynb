**EX.1. Prime Numbers**

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
# Function to check prime number
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Taking input from user
num = int(input("Enter a number: "))

# Checking and printing result
if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is NOT a Prime Number")
```

**Ex.2. Product of Random Numbers**

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random

# Generating two random numbers
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Asking user for the product
print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Enter your answer: "))

# Calculating correct answer it means that it is checking that the multiplication is correct.
correct_answer = num1 * num2

# Checking and displaying result
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print("Incorrect.")
    print(f"The correct answer is: {correct_answer}")
```

**Ex.4. Problem Statement (Question): Hospital Billing System**

A hospital charges its patients based on the following:

* Room charges per day: ₹2,000
* Doctor consultation fee (flat): ₹1,500
* Lab test charges: Based on the number of tests taken. Each test costs ₹300
* Medicine charges: Total cost of medicines provided
* Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied

Write a Python program to calculate the final bill for a patient who:

* Stayed for 4 days
* Had 3 lab tests
* Had medicine charges of ₹2,400


```python
# Hospital Billing System

# Given details
days = 4                     # Patient stayed for 4 days
room_charge_per_day = 2000   # Room charge per day
doctor_fee = 1500            # Fixed doctor consultation fee
lab_tests = 3                # Number of lab tests
cost_per_test = 300          # Cost per lab test
medicine_charges = 2400      # Total medicine cost

# Calculate individual costs
room_charges = days * room_charge_per_day
lab_charges = lab_tests * cost_per_test

# Calculate total before discount
total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if total exceeds 10,000
if total_bill > 10000:
    discount = 0.10 * total_bill   # 10% discount
else:
    discount = 0

# Final bill after discount
final_bill = total_bill - discount

# Display results
print("Room Charges:", room_charges)
print("Doctor Fee:", doctor_fee)
print("Lab Charges:", lab_charges)
print("Medicine Charges:", medicine_charges)
print("Total Bill before Discount:", total_bill)
print("Discount Applied:", discount)
print("Final Bill:", final_bill)
```
