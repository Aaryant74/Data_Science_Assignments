**Task 2: Dimensionality Reduction with PCA:**

1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

3.	Transform the original dataset into the principal components.


```python
# Standardize the Features
# Scaling data so mean = 0 and standard deviation = 1
from sklearn.preprocessing import StandardScaler

# Separate features
X = df.drop('Type', axis=1)

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Applying PCA to reduce dimensionality
from sklearn.decomposition import PCA

# Apply PCA (initially keep all components)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Scree plot helps decide number of important components
import matplotlib.pyplot as plt

plt.plot(pca.explained_variance_ratio_, marker='o')
plt.title("Scree Plot")
plt.xlabel("Number of Components")
plt.ylabel("Explained Variance")
plt.show()
```

```python
# Cumulative Explained Variance
# Choosing number of components that explain ~95% variance
import numpy as np

cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.plot(cumulative_variance, marker='o')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.title("Cumulative Explained Variance")
plt.show()
```

```python
# Reducing dataset to 3 principal components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
# View Transformed Data
X_pca[:5]
```

**Task 3: Clustering with Original Data:**

1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.

2.	Visualize the clustering results using appropriate plots.

3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.


```python
# Applying K-Means clustering to group similar data points
from sklearn.cluster import KMeans

# Choose number of clusters (example: 3)
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit model
clusters = kmeans.fit_predict(X_scaled)
```

```python
# Adding cluster labels to dataset
df['Cluster'] = clusters
df.head()
```

```python
# Visualizing clusters using first two features
import matplotlib.pyplot as plt

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")
plt.show()
```

```python
# Evaluating clustering quality (higher silhouette = better, lower DB index = better)
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_scaled, clusters)
print("Silhouette Score:", sil_score)

from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(X_scaled, clusters)
print("Davies-Bouldin Index:", db_score)
```

**Task 4: Clustering with PCA Data:**

1.	Apply the same clustering algorithm to the PCA-transformed dataset.

2.	Visualize the clustering results obtained from PCA-transformed data.

3.	Compare the clustering results from PCA-transformed data with those from the original dataset.


```python
# Applying K-Means clustering on PCA-transformed data
from sklearn.cluster import KMeans

# Apply K-Means on PCA data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)
```

```python
# Visualizing clusters using first two principal components
import matplotlib.pyplot as plt

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_pca)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Clustering with PCA Data")
plt.show()

# Evaluate PCA Clustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

print("Silhouette Score (PCA):", silhouette_score(X_pca, clusters_pca))
print("Davies-Bouldin Index (PCA):", davies_bouldin_score(X_pca, clusters_pca))
```

```python
# Compare with Original Data
print("Original Silhouette:", silhouette_score(X_scaled, clusters))
print("PCA Silhouette:", silhouette_score(X_pca, clusters_pca))
```

**Task 5: Comparison and Analysis:**

1.	Compare the clustering results obtained from the original dataset and PCA-transformed data.

2.	Discuss any similarities or differences observed in the clustering results.

3.	Reflect on the impact of dimensionality reduction on clustering performance.

4.	Analyze the trade-offs between using PCA and clustering directly on the original dataset.


```python
# Comparing clustering performance of original vs PCA data
import pandas as pd
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Original data metrics
sil_original = silhouette_score(X_scaled, clusters)
db_original = davies_bouldin_score(X_scaled, clusters)

# PCA data metrics
sil_pca = silhouette_score(X_pca, clusters_pca)
db_pca = davies_bouldin_score(X_pca, clusters_pca)

# Create comparison table
comparison = pd.DataFrame({
    'Method': ['Original Data', 'PCA Data'],
    'Silhouette Score': [sil_original, sil_pca],
    'Davies-Bouldin Index': [db_original, db_pca]
})

comparison
```

```python
# Visualizing comparison results
import matplotlib.pyplot as plt

comparison.set_index('Method').plot(kind='bar', figsize=(8,5))
plt.title("Clustering Comparison")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.show()
```

```python
# Analysis:
# PCA reduces dimensionality and helps in better visualization
# Higher silhouette score indicates better clustering
# Lower Davies-Bouldin index indicates better separation
```
