**ESTIMATION AND CONFIDENCE INTERVALS**

**a. Build 99% Confidence Interval Using Sample Standard Deviation**

Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.


```python
import numpy as np
import scipy.stats as stats

# Data
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Sample standard deviation (ddof=1 for sample SD)
std_dev = np.std(data, ddof=1)

# Confidence level
confidence = 0.99
alpha = 1 - confidence

# Degrees of freedom
df = n - 1

# t critical value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Margin of error
margin_error = t_critical * (std_dev / np.sqrt(n))

# Confidence interval
lower = mean - margin_error
upper = mean + margin_error

# Output
print("Sample Size (n):", n)
print("Sample Mean:", round(mean, 3))
print("Sample Standard Deviation:", round(std_dev, 3))
print("t Critical Value:", round(t_critical, 3))
print("Margin of Error:", round(margin_error, 3))

print("\n99% Confidence Interval:")
print(f"({round(lower, 3)}, {round(upper, 3)})")

# Explanation:
# We use the t-distribution because:
# - The sample size is small (n < 30)
# - The population standard deviation is unknown
# - t-distribution gives a more accurate interval for small samples
```

**b. Build 99% Confidence Interval Using Known Population Standard Deviation**

If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for the mean number of characters printed before failure.


```python
import numpy as np
import scipy.stats as stats

# Given data
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])

# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Known population standard deviation
sigma = 0.2   # GIVEN in question

# Confidence level
confidence = 0.99
alpha = 1 - confidence

# Z critical value (since sigma is known)
z_critical = stats.norm.ppf(1 - alpha/2)

# Margin of error
margin_error = z_critical * (sigma / np.sqrt(n))

# Confidence interval
lower = mean - margin_error
upper = mean + margin_error

# Output
print("Sample Mean:", round(mean, 3))
print("Z Critical Value:", round(z_critical, 3))
print("Margin of Error:", round(margin_error, 3))

print("\n99% Confidence Interval:")
print(f"({round(lower, 3)}, {round(upper, 3)})")

# Explanation:
# We used Z-distribution because population standard deviation is known.
# This gives the range where the true mean likely lies.
```

**HYPOTHESIS TESTING**

To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.

Data Provided:

* The theoretical weekly operating cost model: W = $1,000 + $5X
* Sample of 25 restaurants with a mean weekly cost of Rs. 3,050
* Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units

**1. State the Hypotheses statement:**

```python
# Given values
fixed_cost = 1000      # fixed cost
variable_cost = 5      # cost per unit
mean_X = 600           # mean units produced

# Calculate theoretical mean cost
mu_0 = fixed_cost + variable_cost * mean_X

# Print hypotheses
print("Hypotheses Statement:\n")

print("Null Hypothesis (H0):")
print(f"Mean weekly operating cost = {mu_0}")

print("\nAlternative Hypothesis (H1):")
print(f"Mean weekly operating cost ≠ {mu_0}")
```

**2. Calculate the Test Statistic:**

Use the following formula to calculate the test statistic(z):
where:

* ˉxˉ = sample mean weekly cost (Rs. 3,050)
* μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)
* σ = 5*25 units
* n = sample size (25 restaurants)

```python
import numpy as np

# Given values
x_bar = 3050     # sample mean
mu = 4000        # theoretical mean
sigma = 5 * 25   # population standard deviation
n = 25           # sample size

# Step 1: Calculate Z statistic
z = (x_bar - mu) / (sigma / np.sqrt(n))

# Output
print("Z Test Statistic:", z)

# Explanation
# Z shows how far the sample mean is from the theoretical mean.
# A very large value means strong difference.
```

**3. Determine the Probability and compare:**

Using the alpha level of 5% (α = 0.05)


```python
import scipy.stats as stats

# Given Z value
z = -38

# Critical values
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

print("Critical Z values: ±", round(z_critical, 2))

# p-value
p_value = 2 * stats.norm.cdf(z)  # two-tailed

print("p-value:", p_value)
```

**4. Make a Decision:**

Compare the test statistic with the critical value to decide whether to reject the null hypothesis.


```python
# Given values
z = -38
z_critical = 1.96

# Decision
if z < -z_critical or z > z_critical:
    print("Reject the Null Hypothesis (H0)")
else:
    print("Fail to Reject the Null Hypothesis (H0)")
```

**5. Conclusion**

* There is NO strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.
* In fact, the results indicate that the actual operating costs are significantly lower than the predicted cost.
