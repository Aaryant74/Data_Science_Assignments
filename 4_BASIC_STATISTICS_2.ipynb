**Problem Statement : Hospital Patient Data Analysis**


**Context:**
A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. You have two datasets: one with patient info and another with billing details. Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.


**1.	Load the patient dataset and show summary with info().**

```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Load the dataset
df = pd.read_csv('/content/Patient_Data.csv')

# Step 4: Show dataset summary
df.info()
```

**2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].**

```python
# Select specific columns
billing_df = df[['PatientID', 'Department', 'Doctor', 'BillAmount']]

# Display the result
print(billing_df.head())
```

**3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].**

```python
# Drop unnecessary columns
df = df.drop(['ReceptionistID', 'CheckInTime'], axis=1)

# Display updated dataframe
print(df.head())
```

**4.	Use groupby to find total bill amount per department.**

```python
# Group by Department and calculate total BillAmount
dept_total = df.groupby('Department')['BillAmount'].sum()

# Display result
print(dept_total)
```

**5.	Remove duplicate patient records based on PatientID.**

```python
# Remove duplicate records based on PatientID
df = df.drop_duplicates(subset='PatientID')

# Display result
print(df.head())
```

**6.	Fill missing BillAmount values with the mean bill amount.**

```python
# Calculate mean of BillAmount
mean_bill = df['BillAmount'].mean()

# Fill missing values with mean
df['BillAmount'] = df['BillAmount'].fillna(mean_bill)

# Check result
print(df['BillAmount'].head())
```

**7.	Merge the billing dataset with patient dataset on PatientID.**

```python
# Load billing dataset (if not already loaded)
billing_df = pd.read_csv('/content/Billing_Data.csv')

# Merge both datasets on PatientID
merged_df = pd.merge(df, billing_df, on='PatientID')

# Display result
print(merged_df.head())
```

**8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).**

```python
# Example: New patients dataframe (you can replace with your actual data)
new_patients = pd.DataFrame({
    'PatientID': [101, 102],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Sharma', 'Dr. Mehta'],
    'BillAmount': [5000, 7000]
})

# Concatenate row-wise
df = pd.concat([df, new_patients], axis=0, ignore_index=True)

# Display result
print(df.tail())
```

**9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).**

```python
# Example: New billing columns DataFrame
new_columns = pd.DataFrame({
    'InsuranceCovered': [True, False, True, False],
    'FinalAmount': [4500, 7000, 3000, 6500]
})

# Concatenate column-wise
df = pd.concat([df, new_columns], axis=1)

# Display result
print(df.head())
```
