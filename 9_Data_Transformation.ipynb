●	Apply scaling techniques to numerical features:

a.	Standard Scaling   b. Min-Max Scaling

●	Discuss the scenarios where each scaling technique is preferred and why.


```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/adult_with_headers.csv')
df.head()

# Handling Missing Values
# Replace ? with NaN
df.replace(' ?', np.nan, inplace=True)

# Check missing values
df.isnull().sum()

# Drop missing values
df = df.dropna()

# Scaling Numerical Features
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standard = df.copy()
df_standard[num_cols] = scaler.fit_transform(df_standard[num_cols])

df_standard.head()

# MinMax Scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = scaler.fit_transform(df_minmax[num_cols])

df_minmax.head()
```

**Discussing When to Use Which Scaling :**

1. Standard Scaling (Z-score)
When data has outliers, Used in Linear Regression, Logistic Regression, SVM, Makes data mean = 0 and std = 1.

Best for normal distribution

2. Min-Max Scaling
When you need values between 0 and 1, Used in Neural Networks, KNN, Sensitive to outliers.

Best when no extreme values

**2. Encoding Techniques:**

* Apply One-Hot Encoding to categorical variables with less than 5 categories.
* Use Label Encoding for categorical variables. Data Exploration and Preprocessing:
* Load the dataset and conduct basic data exploration (summary statistics, missing values, data types).
* les with more than 5 categories.
* Discuss the pros and cons of One-Hot Encoding and Label Encoding.


```python
# Basic Data Exploration
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary Stats:\n", df.describe())

df.head()

cat_cols = df.select_dtypes(include='object').columns

# Check number of unique values
for col in cat_cols:
    print(col, ":", df[col].nunique())

# One Hot Encoding in less than 5 Categories
low_card_cols = [col for col in cat_cols if df[col].nunique() < 5]

df_ohe = pd.get_dummies(df, columns=low_card_cols, drop_first=True)

df_ohe.head()

# Labelling Encoding in more than 5 Categories
from sklearn.preprocessing import LabelEncoder

high_card_cols = [col for col in cat_cols if df[col].nunique() >= 5]

le = LabelEncoder()

for col in high_card_cols:
    df_ohe[col] = le.fit_transform(df_ohe[col])

df_ohe.head()

print("Final Shape:", df_ohe.shape)
```

***Pros & Cons:***

**One-Hot Encoding**

Pros:
No order/relationship created
Good for small categories

Cons:
Creates many columns
Increases dataset size


**Label Encoding**

Pros:

Simple and fast
No extra columns

Cons:

Creates false order (e.g., 0 < 1 < 2)
Can mislead model
